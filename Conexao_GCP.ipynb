Crie listas com 5 valores cada, com os respectivos nomes:
- matricula
- nome
- nota1
- nota2
- nota3
Os valores devem fazer sentido com cada lista.

Crie um dicionário a partir dessas listas com chave para cada lista.
Crie um dataframe com esse dicionário.
Gere um arquivo csv chamado notas_alunos.csv, suba para o storage em dfs_bruto
No dataframe, adicione uma coluna que realize a media das 3 notas
Gere um arquivo csv com o dataframe tratado chamado notas_alunos_finais.csv suba para o storagem em dfs_tratados

Entregue os códigos de forma individual.

```python
# criando as listas
matriculas = [12223, 33423, 44323, 53465, 12221]
nomes = [
    'Sukama Sarmento', 'Marcela Perove', 'Carla Kayembe', 'Jaime Jorge', 'Daniel Victor'
]
notas1 = [9, 8, 9.5, 6, 7.8]
notas2 = [5, 10, 9.8, 8, 10]
notas3 = [1, 3.3, 4.3, 9, 8]
# criando o dicionário
dicionario_alunos = {
    'matricula' : matriculas,
    'nome' : nomes,
    'nota1' : notas1,
    'nota2' : notas2,
    'nota3' : notas3
}
```

```python
# criando a DataFrame
import pandas as pd
df_alunos = pd.DataFrame(dicionario_alunos)
display(df_alunos)
```

```python
# Gerando o arquivo .csv
df_alunos.to_csv('notas_alunos.csv', index=False)
```

```python
# importando arquivo do cloud storage
df_alunos2 = pd.read_csv('https://storage.googleapis.com/sukama-bucket/dfs_brutos/notas_alunos.csv')
display(df_alunos2)
```

```python
# adicionando a coluna média ao Dataframe
df_alunos['media'] = round((df_alunos['nota1'] + df_alunos['nota2'] + df_alunos['nota3']) / 3, 2)
display(df_alunos)
```

```python
# gerando arquivo notas_alunos_final.csv
df_alunos.to_csv('notas_alunos_final.csv', index=False)
```
